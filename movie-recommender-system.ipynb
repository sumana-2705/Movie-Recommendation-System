# Movie Recommendation System 


```python
import numpy as np 
import pandas as pd
import ast
```

```python
# importing Datasets

movies = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
credits = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
# Merging the both datasets 
movies = movies.merge(credits, on = 'title')
print(movies.shape)
```

```python
movies.info()
```

```python
# Creating a new dataset with required columns
movies = pd.DataFrame(movies[['movie_id', 'title', 'overview', 'genres', 'keywords','cast','crew']])
movies.head()
```

## Data Pre-Processing

```python
movies.isnull().sum()
```

```python
# removing empty rows
movies.dropna(inplace = True)

movies.isnull().sum()
```

```python
# check for repitition
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        
    return L
    
```

Original: '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

Modified: ['Action', 'Adventure', 'Fantasy', 'Scifi']

```python
movies['genres'] =  movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_cast(obj):
    L = []
    cast = 0
    for i in ast.literal_eval(obj):
        if cast != 3:
            L.append(i['name'])
            cast += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head()
```

```python
# converting 'Overview' column from a string to list 
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Removing spaces between to avoid confusions 
# 'Sam Mendes' --> 'SamMendes'

movies['genres'] = movies['genres'].apply (lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply (lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply (lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply (lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# Creating a new column 'tag' for recommendations
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# new dataframe named 'new_df' for our later steps for recommendation
new_df = pd.DataFrame(movies[['movie_id', 'title', 'tag']])
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply (lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tag'][0]
```

```python
new_df['tag']= new_df['tag'].apply(lambda x:x.lower())

new_df['tag'][0]
```

```python
new_df['tag'][1]
```

## Building a Model

```python
# We need to establish a relation between these tag lines based on the text
# converting text to vector using bagging
# 1. extracting common 5000 words
# 2. removing stop words like is, in, are etc..,

from sklearn.feature_extraction.text import CountVectorizer

cv= CountVectorizer (max_features = 5000, stop_words = 'english')


```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# Stemming the words in most common 5000 words
# Stemming = extracting the original form (noun, verb) from the words
# Example: [loving, loves, loved] --> [love, love, love]

import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
print(ps.stem('caring'))
print (ps.stem('dancing'))
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ". join(y)
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
new_df['tag'][0]
```

#### We are not interested in calculating the distance between the vectors, but we will calculate the angle between them to estimate the similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity [1]
```

```python
def recommend (movie):
    #index fetching
    movie_index = new_df[new_df['title']== movie].index[0]
    
    # calculating distances
    distances = similarity [movie_index]
    # sorting the distances
    movies_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print (new_df.iloc[i[0]].title)
    
```

```python
new_df[new_df['title']== 'Avatar'].index
```

```python
recommend('Batman')
```
